```python
# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║   SMART WATER METER — SOLUTION 4 (FINAL)                                    ║
# ║   PatchTST + BiLSTM-Attn + Ridge OOF Stacking                               ║
# ║   Dual Active/Quiet Models + Adaptive TOD Clamp + Extended CAS               ║
# ║   CORRECTED: LPM·h × 12 = litres  |  LPM·5min × 1 = litres                 ║
# ╚══════════════════════════════════════════════════════════════════════════════╝


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL 1 — INSTALLATION  (run once → Runtime > Restart Runtime)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
"""
!pip install -q torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118
!pip install -q transformers==4.40.0
!pip install -q scikit-learn joblib pandas numpy matplotlib seaborn tqdm einops

import torch
print(f"PyTorch : {torch.__version__}")
print(f"CUDA    : {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU     : {torch.cuda.get_device_name(0)}")
"""


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL 2 — IMPORTS · CONFIG · PREPROCESSING
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

import os, json, warnings, random, joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from pathlib import Path
from scipy.interpolate import CubicSpline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import KFold
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, Subset
from transformers import PatchTSTConfig, PatchTSTForPrediction

warnings.filterwarnings("ignore")

SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"[INFO] Running on: {DEVICE}")

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# GLOBAL CONFIGURATION
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
CONTEXT_LEN     = 168
PRED_LEN        = 24
BATCH_SIZE      = 32
EPOCHS          = 100
LR              = 1e-3
TRAIN_RATIO     = 0.70
VAL_RATIO       = 0.15
OOF_FOLDS       = 5
RIDGE_ALPHA     = 1.0
OOF_EPOCHS      = 60

SHORT_GAP_PROB  = 0.004
LONG_GAP_PROB   = 0.0006
LONG_GAP_MIN_HR = 4
LONG_GAP_MAX_HR = 8

TOD_CLAMP_QUANTILE = 0.90

# ── DUAL-MODEL HOUR SPLIT ──────────────────────────────────────────────────────
ACTIVE_HOURS       = set(range(6, 22))
QUIET_HOURS        = set(range(22, 24)) | set(range(0, 6))
BOUNDARY_BLEND_HRS = 2
# ──────────────────────────────────────────────────────────────────────────────

SAVE_DIR_DUAL = Path("./smart_meter_models_dual")

FEATURE_COLS = [
    "flowrate",
    "hour_sin", "hour_cos",
    "day_sin",  "day_cos",
    "month_sin","month_cos",
    "lag_24h",
    "lag_168h",
]
N_FEATURES = len(FEATURE_COLS)   # 9

# ── SOLUTION 4 CONFIGURATION ──────────────────────────────────────────────────
CAS_PRE_SCAN_HRS    = 12
CAS_POST_SCAN_HRS   = 12
CAS_MIN_ACTIVE_OBS  = 2
ACT_SCALE_THRESHOLD = 0.15
CAS_SCALE_FLOOR     = 0.01
CAS_SCALE_CEIL      = 3.0
CAS_TAPER_HRS       = 3
TRANSITION_HOURS    = {4, 5, 21, 22}
TRANSITION_QUANTILE = 0.50
NORMAL_QUANTILE     = 0.90
CAS_VERBOSE         = True

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# UNIT CONVERSION CONSTANTS
# Hourly value = mean of 12 five-minute readings
#   Total litres (hourly) = LPM_hourly_mean × 12
#   Total litres (5-min)  = LPM_5min_value  × 1
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
LPM_H_TO_LITRES    = 12   # multiply hourly LPM·h value by 12 to get litres
LPM_5MIN_TO_LITRES =  1   # multiply 5-min LPM value by 1  to get litres


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# PREPROCESSING FUNCTIONS
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def load_raw_data(filepath: str) -> pd.DataFrame:
    df = pd.read_csv(filepath, parse_dates=["timestamp"])
    if "flow_rate_lpm" in df.columns:
        df.rename(columns={"flow_rate_lpm": "flowrate"}, inplace=True)
    df.set_index("timestamp", inplace=True)
    df.sort_index(inplace=True)
    full_idx = pd.date_range(df.index.min(), df.index.max(), freq="5min")
    df = df.reindex(full_idx); df.index.name = "timestamp"
    print(f"[LOAD]  {len(df):,} rows  |  "
          f"{df.index.min()} → {df.index.max()}")
    print(f"[LOAD]  NaN: {df['flowrate'].isna().sum():,}  |  "
          f"Zeros: {(df['flowrate']==0).sum():,} "
          f"({(df['flowrate']==0).sum()/len(df)*100:.1f}%)")
    return df


def inject_artificial_missing(df: pd.DataFrame, seed: int = SEED) -> tuple:
    rng = np.random.default_rng(seed); n = len(df)
    df_c = df.copy(); mask = np.zeros(n, dtype=bool)
    short_mask = rng.random(n) < SHORT_GAP_PROB; mask |= short_mask
    long_starts, long_ends = [], []
    i = 0
    while i < n:
        if rng.random() < LONG_GAP_PROB:
            glen  = int(rng.integers(LONG_GAP_MIN_HR*12, LONG_GAP_MAX_HR*12+1))
            end_i = min(i+glen, n); mask[i:end_i] = True
            long_starts.append(df_c.index[i])
            long_ends.append(df_c.index[end_i-1]); i = end_i
        else: i += 1
    df_c.loc[df_c.index[mask], "flowrate"] = np.nan
    pct = mask.sum()/n*100
    print(f"\n[INJECT]  {mask.sum():,} NaN ({pct:.2f}%)  "
          f"| Short: {short_mask.sum():,}  Long: {len(long_starts)}")
    if long_starts:
        print(f"[INJECT]  Example: {long_starts[0]} → {long_ends[0]}")
    return df_c, dict(mask=mask, short_gap_count=int(short_mask.sum()),
                      long_gap_starts=long_starts, long_gap_ends=long_ends)


def interpolate_short_gaps(df: pd.DataFrame, max_gap_pts: int = 3) -> pd.DataFrame:
    df_out = df.copy(); vals = df_out["flowrate"].values.copy(); n = len(vals)
    runs, in_gap, gs = [], False, 0
    for i in range(n):
        if np.isnan(vals[i]):
            if not in_gap: in_gap, gs = True, i
        else:
            if in_gap: runs.append((gs, i-1)); in_gap = False
    if in_gap: runs.append((gs, n-1))
    filled = 0
    for (s, e) in runs:
        if (e-s+1) <= max_gap_pts:
            l, r = s-1, e+1
            if l>=0 and r<n and not np.isnan(vals[l]) and not np.isnan(vals[r]):
                cs = CubicSpline([l,r],[vals[l],vals[r]])
                for idx in range(s,e+1): vals[idx] = max(0.0, float(cs(idx)))
            else: vals[s:e+1] = 0.0
            filled += (e-s+1)
    df_out["flowrate"] = vals
    print(f"[INTERP]  Filled {filled:,}  |  "
          f"NaN remaining: {int(np.isnan(vals).sum()):,}")
    return df_out


def downsample_hourly(df: pd.DataFrame, min_pts: int = 6) -> pd.DataFrame:
    hourly = df["flowrate"].resample("1h").apply(
        lambda x: x.mean() if x.count() >= min_pts else np.nan
    ).to_frame(); hourly.index.name = "timestamp"
    print(f"\n[RESAMP]  {len(hourly):,} hourly rows  |  "
          f"NaN: {hourly['flowrate'].isna().sum()}")
    return hourly


def add_cyclical_features(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy(); h, dw, mo = df.index.hour, df.index.dayofweek, df.index.month
    df["hour_sin"]  = np.sin(2*np.pi*h/24);  df["hour_cos"]  = np.cos(2*np.pi*h/24)
    df["day_sin"]   = np.sin(2*np.pi*dw/7);  df["day_cos"]   = np.cos(2*np.pi*dw/7)
    df["month_sin"] = np.sin(2*np.pi*mo/12); df["month_cos"] = np.cos(2*np.pi*mo/12)
    return df


class Log1pStandardScaler:
    def __init__(self): self.sc = StandardScaler()
    def _lp(self, s):
        return np.where(np.isnan(s.values), np.nan,
                        np.log1p(np.maximum(s.values, 0.0)))
    def fit_transform(self, s):
        lv = self._lp(s); self.sc.fit(lv[~np.isnan(lv)].reshape(-1,1))
        return pd.Series(np.where(np.isnan(lv), np.nan,
                                  self.sc.transform(lv.reshape(-1,1)).flatten()),
                         index=s.index)
    def transform(self, s):
        lv = self._lp(s)
        return pd.Series(np.where(np.isnan(lv), np.nan,
                                  self.sc.transform(lv.reshape(-1,1)).flatten()),
                         index=s.index)
    def inverse_transform(self, arr):
        arr = np.asarray(arr, dtype=np.float64)
        if arr.size == 0: return arr.astype(np.float32)
        return np.maximum(
            np.expm1(self.sc.inverse_transform(arr.reshape(-1,1)).flatten()),
            0.0).astype(np.float32)
    @property
    def mean_(self): return float(self.sc.mean_[0])
    @property
    def scale_(self): return float(self.sc.scale_[0])


class WaterMeterDataset(Dataset):
    def __init__(self, df, context_len, pred_len, skip_nan=True):
        self.context_len, self.pred_len = context_len, pred_len
        df_f = df[FEATURE_COLS].copy().ffill().fillna(0.0)
        feat = df_f.values.astype(np.float32)
        raw  = df["flowrate"].values
        self.samples, self.start_times = [], []
        for i in range(len(feat)-context_len-pred_len+1):
            x = feat[i:i+context_len]; y = raw[i+context_len:i+context_len+pred_len]
            if skip_nan and np.isnan(y).any(): continue
            self.samples.append((x, y.astype(np.float32)))
            self.start_times.append(df.index[i+context_len])
        print(f"[DS]      {len(self.samples):,} windows  "
              f"(ctx={context_len}, pred={pred_len}, feats={N_FEATURES})")
    def __len__(self): return len(self.samples)
    def __getitem__(self, i):
        x, y = self.samples[i]; return torch.tensor(x), torch.tensor(y)


def build_tod_cap_table(df_clean_hourly: pd.DataFrame,
                         quantile: float = TOD_CLAMP_QUANTILE) -> dict:
    caps = {}
    flow = df_clean_hourly["flowrate"].dropna()
    for h in range(24):
        hour_vals = flow[flow.index.hour == h].values
        if len(hour_vals) > 0:
            caps[h] = float(np.quantile(hour_vals, quantile))
        else:
            caps[h] = float("inf")
    print(f"\n[TOD-CAP]  Built {quantile*100:.0f}th-percentile caps "
          f"from {len(flow):,} clean hourly readings")
    return caps


def build_pipeline(filepath, context_len=CONTEXT_LEN, pred_len=PRED_LEN,
                   train_ratio=TRAIN_RATIO, val_ratio=VAL_RATIO,
                   deploy_mode=False) -> dict:
    sep = "="*68
    print(sep); print(f"  PREPROCESSING  (deploy_mode={deploy_mode})"); print(sep)
    df_raw = load_raw_data(filepath)

    _ch = downsample_hourly(df_raw)
    _ch = add_cyclical_features(_ch)
    _ch["lag_24h"]  = _ch["flowrate"].shift(24).ffill().fillna(0.0)
    _ch["lag_168h"] = _ch["flowrate"].shift(168).ffill().fillna(0.0)
    df_clean_hourly = _ch

    tod_caps = build_tod_cap_table(df_clean_hourly, quantile=TOD_CLAMP_QUANTILE)

    inject_meta = dict(long_gap_starts=[], long_gap_ends=[],
                       mask=np.zeros(len(df_raw), dtype=bool),
                       short_gap_count=0)
    if deploy_mode:
        print("\n[MODE]  deploy_mode=True — using real gaps only")
        df_work = df_raw.copy()
    else:
        print("\n[INJECT]  Injecting artificial gaps for validation...")
        df_work, inject_meta = inject_artificial_missing(df_raw)

    df_i = interpolate_short_gaps(df_work)
    df_h = downsample_hourly(df_i)
    df_f = add_cyclical_features(df_h)

    n  = len(df_f); s1 = int(n*train_ratio); s2 = int(n*(train_ratio+val_ratio))
    df_tr = df_f.iloc[:s1].copy()
    df_va = df_f.iloc[s1:s2].copy()
    df_te = df_f.iloc[s2:].copy()
    print(f"\n[SPLIT]   Train={len(df_tr):,} h  "
          f"Val={len(df_va):,} h  Test={len(df_te):,} h")

    scaler = Log1pStandardScaler()
    df_tr["flowrate"] = scaler.fit_transform(df_tr["flowrate"])
    df_va["flowrate"] = scaler.transform(df_va["flowrate"])
    df_te["flowrate"] = scaler.transform(df_te["flowrate"])
    print(f"[SCALE]   log1p-Z  μ={scaler.mean_:.5f}  σ={scaler.scale_:.5f}")

    for df_part in [df_tr, df_va, df_te]:
        df_part["lag_24h"]  = df_part["flowrate"].shift(24).ffill().fillna(0.0)
        df_part["lag_168h"] = df_part["flowrate"].shift(168).ffill().fillna(0.0)
    print("[FEAT]    lag_24h + lag_168h added")

    print("[TRAIN]   ", end="")
    tr_ds = WaterMeterDataset(df_tr, context_len, pred_len)
    print("[VAL]     ", end="")
    va_ds = WaterMeterDataset(df_va, context_len, pred_len)
    print("[TEST]    ", end="")
    te_ds = WaterMeterDataset(df_te, context_len, pred_len)

    tr_dl = DataLoader(tr_ds, BATCH_SIZE, shuffle=True,  drop_last=True,  num_workers=0, pin_memory=True)
    va_dl = DataLoader(va_ds, BATCH_SIZE, shuffle=False, drop_last=False, num_workers=0, pin_memory=True)
    te_dl = DataLoader(te_ds, BATCH_SIZE, shuffle=False, drop_last=False, num_workers=0, pin_memory=True)
    print(f"[LOADER]  Train={len(tr_dl)}  Val={len(va_dl)}  Test={len(te_dl)}")

    return dict(df_raw=df_raw, df_corrupt=df_work,
                df_interpolated=df_i, df_hourly=df_h,
                df_clean_hourly=df_clean_hourly,
                df_train=df_tr, df_val=df_va, df_test=df_te,
                scaler=scaler, tod_caps=tod_caps,
                train_loader=tr_dl, val_loader=va_dl, test_loader=te_dl,
                train_dataset=tr_ds, val_dataset=va_ds, test_dataset=te_ds,
                inject_meta=inject_meta,
                num_features=N_FEATURES,
                context_len=context_len, pred_len=pred_len)


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL 3 — MODELS
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

class TemporalAttention(nn.Module):
    def __init__(self, h):
        super().__init__(); self.a = nn.Linear(h, 1, bias=False)
    def forward(self, x):
        w = F.softmax(self.a(x).squeeze(-1), dim=-1)
        return (w.unsqueeze(-1) * x).sum(dim=1)


class BiLSTMForecaster(nn.Module):
    def __init__(self, num_features=N_FEATURES, hidden_size=64,
                 num_layers=2, dropout=0.3, pred_len=PRED_LEN):
        super().__init__()
        self.pred_len = pred_len
        self.lstm = nn.LSTM(num_features, hidden_size, num_layers,
                            dropout=dropout if num_layers>1 else 0.0,
                            batch_first=True, bidirectional=True)
        self.attn = TemporalAttention(hidden_size*2)
        self.norm = nn.LayerNorm(hidden_size*2)
        self.drop = nn.Dropout(dropout)
        self.fc   = nn.Linear(hidden_size*2, pred_len)
        for nm, p in self.lstm.named_parameters():
            if   "weight_ih" in nm: nn.init.xavier_uniform_(p)
            elif "weight_hh" in nm: nn.init.orthogonal_(p)
            elif "bias"       in nm: nn.init.zeros_(p)
        nn.init.xavier_uniform_(self.fc.weight)
        nn.init.zeros_(self.fc.bias)
    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(self.drop(self.norm(self.attn(out))))


class PatchTSTForecaster(nn.Module):
    def __init__(self, context_len=CONTEXT_LEN, pred_len=PRED_LEN,
                 patch_length=16, d_model=128, num_heads=4,
                 num_layers=3, dropout=0.2):
        super().__init__()
        self.pred_len = pred_len
        cfg = PatchTSTConfig(
            num_input_channels=1, context_length=context_len,
            prediction_length=pred_len, patch_length=patch_length,
            patch_stride=patch_length//2, d_model=d_model,
            num_attention_heads=num_heads, num_hidden_layers=num_layers,
            ffn_dim=d_model*4, dropout=dropout, head_dropout=dropout,
            distribution_output="normal", loss="mse",
            do_mask_input=False, channel_attention=False, use_cls_token=False)
        self.model = PatchTSTForPrediction(cfg)
    def forward(self, x):
        return self.model(
            past_values=x[:,:,0:1],
            past_time_features=x[:,:,1:]
        ).prediction_outputs.squeeze(-1)


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL 4 — TRAINING · METRICS · OOF · ENSEMBLE
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def smape(y_true, y_pred):
    return float(np.mean(
        2*np.abs(y_true-y_pred) /
        (np.abs(y_true)+np.abs(y_pred)+1e-6)) * 100)


def compute_metrics(y_true, y_pred, label=""):
    y_true = np.asarray(y_true, dtype=np.float64)
    y_pred = np.asarray(y_pred, dtype=np.float64)
    rmse = float(np.sqrt(mean_squared_error(y_true, y_pred)))
    mae  = float(mean_absolute_error(y_true, y_pred))
    sm   = smape(y_true, y_pred)
    ss   = float(np.sum((y_true - y_true.mean())**2))
    r2   = float(r2_score(y_true, y_pred)) if (len(y_true)>=5 and ss>=1e-8) else float("nan")
    r2s  = f"{r2:.4f}" if not np.isnan(r2) else "N/A"
    tag  = f"[{label}]" if label else ""
    print(f"  {tag:<24}  RMSE={rmse:.5f}  MAE={mae:.5f}  "
          f"R²={r2s}  sMAPE={sm:.1f}%")
    return dict(rmse=rmse, mae=mae, r2=r2, smape=sm)


def train_model(model, tr_dl, va_dl, epochs=EPOCHS, lr=LR,
                name="Model", patience=20):
    model.to(DEVICE)
    crit  = nn.HuberLoss(delta=1.0)
    opt   = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    cos_s = torch.optim.lr_scheduler.CosineAnnealingLR(
        opt, T_max=epochs, eta_min=lr*0.005)
    plt_s = torch.optim.lr_scheduler.ReduceLROnPlateau(
        opt, "min", factor=0.5, patience=6, min_lr=lr*0.005)
    best_val, best_state, no_imp = float("inf"), None, 0
    tr_h, va_h = [], []
    n_p = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"\n{'='*65}\n  Training  {name}  |  "
          f"Params: {n_p:,}  |  {DEVICE}\n{'='*65}")
    for ep in range(1, epochs+1):
        model.train(); tl = 0.0
        for xb, yb in tr_dl:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            opt.zero_grad(set_to_none=True)
            loss = crit(model(xb), yb)
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            opt.step(); tl += loss.item()
        tl /= len(tr_dl)
        model.eval(); vl = 0.0
        with torch.no_grad():
            for xb, yb in va_dl:
                vl += crit(model(xb.to(DEVICE)), yb.to(DEVICE)).item()
        vl /= len(va_dl)
        cos_s.step(); plt_s.step(vl)
        tr_h.append(tl); va_h.append(vl)
        if ep % 10 == 0 or ep == 1:
            print(f"  Ep {ep:03d}/{epochs}  Train={tl:.5f}  "
                  f"Val={vl:.5f}  LR={opt.param_groups[0]['lr']:.2e}")
        if vl < best_val - 1e-6:
            best_val  = vl
            best_state= {k:v.clone() for k,v in model.state_dict().items()}
            no_imp    = 0
        else:
            no_imp += 1
            if no_imp >= patience:
                print(f"  ↳ Early stop ep {ep}  "
                      f"(best val={best_val:.5f})"); break
    if best_state: model.load_state_dict(best_state)
    print(f"  ✓ Done. Best val loss: {best_val:.5f}")
    return tr_h, va_h


@torch.no_grad()
def collect_preds(model, loader):
    model.eval().to(DEVICE); P, T = [], []
    for xb, yb in loader:
        P.append(model(xb.to(DEVICE)).cpu().numpy())
        T.append(yb.numpy())
    return np.concatenate(P).flatten(), np.concatenate(T).flatten()


def make_loader(ds, idx, shuffle=False):
    return DataLoader(Subset(ds, idx.tolist()), BATCH_SIZE,
                      shuffle=shuffle, drop_last=False,
                      num_workers=0, pin_memory=True)


def generate_oof_preds(model_cls, model_kwargs, train_ds,
                       n_folds=OOF_FOLDS, lr=LR,
                       epochs=EPOCHS, name="Model"):
    n  = len(train_ds); kf = KFold(n_splits=n_folds, shuffle=False)
    oof_p = np.zeros((n, PRED_LEN), dtype=np.float32)
    oof_t = np.zeros((n, PRED_LEN), dtype=np.float32)
    print(f"\n[OOF]  {n_folds}-fold OOF for {name}")
    for fold, (tr_idx, va_idx) in enumerate(kf.split(np.arange(n))):
        m = model_cls(**model_kwargs)
        train_model(m, make_loader(train_ds, tr_idx, True),
                    make_loader(train_ds, va_idx),
                    epochs=min(epochs, OOF_EPOCHS), lr=lr,
                    name=f"{name} fold {fold+1}/{n_folds}", patience=12)
        m.eval().to(DEVICE); P, T = [], []
        with torch.no_grad():
            for xb, yb in make_loader(train_ds, va_idx):
                P.append(m(xb.to(DEVICE)).cpu().numpy())
                T.append(yb.numpy())
        pd2 = np.concatenate(P); td2 = np.concatenate(T)
        oof_p[va_idx] = pd2; oof_t[va_idx] = td2
        print(f"  Fold {fold+1}/{n_folds}  windows={len(va_idx)} ✓")
    return oof_p, oof_t


def build_smart_ensemble(bilstm, patchtst, oof_bl, oof_pt, oof_y,
                          val_bl, val_pt, val_y, label=""):
    bl_flat = oof_bl.flatten(); pt_flat = oof_pt.flatten()
    y_flat  = oof_y.flatten()
    bl_r2   = r2_score(y_flat, bl_flat)
    pt_r2   = r2_score(y_flat, pt_flat)
    tag = f" [{label}]" if label else ""
    print(f"\n[SMART-ENS{tag}]  OOF R²  "
          f"BiLSTM={bl_r2:.4f}  PatchTST={pt_r2:.4f}")
    if pt_r2 < bl_r2 - 0.02:
        print(f"[SMART-ENS{tag}]  PatchTST underperforms → BiLSTM-ONLY blend")
        use_patchtst = False
        X_tr = bl_flat.reshape(-1,1)
    else:
        print(f"[SMART-ENS{tag}]  Both models contribute → 2-model Ridge blend")
        use_patchtst = True
        X_tr = np.column_stack([bl_flat, pt_flat])
    ridge = Ridge(alpha=RIDGE_ALPHA)
    ridge.fit(X_tr, y_flat)
    w = ridge.coef_; b = float(np.squeeze(ridge.intercept_))
    if use_patchtst:
        print(f"[SMART-ENS{tag}]  Ridge: {w[0]:.4f}×BiLSTM "
              f"+ {w[1]:.4f}×PatchTST + {b:.4f}")
    else:
        print(f"[SMART-ENS{tag}]  Ridge (BiLSTM-only): "
              f"{w[0]:.4f}×BiLSTM + {b:.4f}")
    return ridge, use_patchtst


def plot_loss_curves(tr, va, name=""):
    fig, ax = plt.subplots(figsize=(10,4))
    ax.plot(tr, label="Train", color="#1976D2", lw=2)
    ax.plot(va, label="Val",   color="#E53935", lw=2)
    ax.set_title(f"{name} — Loss Curve", fontweight="bold")
    ax.set_xlabel("Epoch"); ax.set_ylabel("Huber Loss")
    ax.legend(); ax.grid(alpha=0.3); plt.tight_layout(); plt.show()


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL 4-C — DUAL-MODEL DATASET SPLIT
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def split_dataset_by_hour(ds: WaterMeterDataset,
                           active_hours: set = ACTIVE_HOURS,
                           quiet_hours:  set = QUIET_HOURS) -> tuple:
    active_list, quiet_list = [], []
    for i, st in enumerate(ds.start_times):
        if st.hour in active_hours:
            active_list.append(i)
        else:
            quiet_list.append(i)

    active_idx = np.array(active_list, dtype=np.int64)
    quiet_idx  = np.array(quiet_list,  dtype=np.int64)
    n_total  = len(ds)
    n_active = len(active_idx)
    n_quiet  = len(quiet_idx)

    print(f"\n[DUAL-SPLIT]  Total windows : {n_total:,}")
    print(f"[DUAL-SPLIT]  Active windows: {n_active:,}  "
          f"({n_active/n_total*100:.1f}%)  hours {sorted(active_hours)}")
    print(f"[DUAL-SPLIT]  Quiet windows : {n_quiet:,}  "
          f"({n_quiet/n_total*100:.1f}%)  hours {sorted(quiet_hours)}")
    if n_quiet < 100:
        print(f"[DUAL-SPLIT]  ⚠ WARNING: only {n_quiet} quiet windows — "
              f"PatchTST may overfit. OOF folds reduced to 3.")
    return active_idx, quiet_idx


def _build_split_loaders(ds, idx, val_split=0.15):
    n     = len(idx)
    n_val = max(1, int(n * val_split))
    tr_i  = idx[:n - n_val]
    va_i  = idx[n - n_val:]
    tr_dl = DataLoader(Subset(ds, tr_i.tolist()), BATCH_SIZE,
                       shuffle=True, drop_last=True, num_workers=0, pin_memory=True)
    va_dl = DataLoader(Subset(ds, va_i.tolist()), BATCH_SIZE,
                       shuffle=False, drop_last=False, num_workers=0, pin_memory=True)
    return tr_dl, va_dl, tr_i, va_i


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL 4-D — TRAIN DUAL MODELS
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def train_dual_models(pipe: dict,
                       epochs: int = EPOCHS,
                       use_oof: bool = True) -> dict:
    tr_ds     = pipe["train_dataset"]
    va_dl     = pipe["val_loader"]
    te_dl     = pipe["test_loader"]
    scaler    = pipe["scaler"]
    ctx_len   = pipe["context_len"]
    pred_len  = pipe["pred_len"]
    n_feat    = pipe["num_features"]

    active_idx, quiet_idx = split_dataset_by_hour(tr_ds)
    n_quiet     = len(quiet_idx)
    quiet_folds = min(OOF_FOLDS, max(3, n_quiet // 200))
    print(f"[DUAL]  Quiet OOF folds: {quiet_folds}")

    sep = "="*68

    # ── ACTIVE MODEL ──────────────────────────────────────────────────────────
    print(f"\n{sep}\n  ACTIVE MODEL  (hours {sorted(ACTIVE_HOURS)})\n{sep}")
    tr_dl_a, va_dl_a, _, _ = _build_split_loaders(tr_ds, active_idx)
    print(f"[ACTIVE]  Train batches={len(tr_dl_a)}  Val batches={len(va_dl_a)}")

    bilstm_a  = BiLSTMForecaster(num_features=n_feat, pred_len=pred_len)
    trh_a, vah_a = train_model(bilstm_a, tr_dl_a, va_dl_a,
                                epochs=epochs, lr=LR,
                                name="BiLSTM-Attn [ACTIVE]")
    plot_loss_curves(trh_a, vah_a, "BiLSTM-Attn [ACTIVE]")

    patchtst_a = PatchTSTForecaster(context_len=ctx_len, pred_len=pred_len)
    trh_a2, vah_a2 = train_model(patchtst_a, tr_dl_a, va_dl_a,
                                  epochs=epochs, lr=LR*0.5,
                                  name="PatchTST [ACTIVE]")
    plot_loss_curves(trh_a2, vah_a2, "PatchTST [ACTIVE]")

    if use_oof:
        bl_oof_a, y_oof_a = generate_oof_preds(
            BiLSTMForecaster, dict(num_features=n_feat, pred_len=pred_len),
            tr_ds, n_folds=OOF_FOLDS, lr=LR, epochs=OOF_EPOCHS,
            name="BiLSTM [ACTIVE]")
        pt_oof_a, _ = generate_oof_preds(
            PatchTSTForecaster, dict(context_len=ctx_len, pred_len=pred_len),
            tr_ds, n_folds=OOF_FOLDS, lr=LR*0.5, epochs=OOF_EPOCHS,
            name="PatchTST [ACTIVE]")
    else:
        bl_oof_a, y_oof_a = generate_oof_preds(
            BiLSTMForecaster, dict(num_features=n_feat, pred_len=pred_len),
            tr_ds, n_folds=1, lr=LR, epochs=30, name="BiLSTM [ACTIVE]")
        pt_oof_a, _ = generate_oof_preds(
            PatchTSTForecaster, dict(context_len=ctx_len, pred_len=pred_len),
            tr_ds, n_folds=1, lr=LR*0.5, epochs=30, name="PatchTST [ACTIVE]")

    bl_va_a, y_va_a = collect_preds(bilstm_a,   va_dl)
    pt_va_a, _      = collect_preds(patchtst_a, va_dl)
    meta_a, use_pt_a = build_smart_ensemble(
        bilstm_a, patchtst_a,
        bl_oof_a, pt_oof_a, y_oof_a,
        bl_va_a,  pt_va_a,  y_va_a,
        label="ACTIVE")

    # ── QUIET MODEL ───────────────────────────────────────────────────────────
    print(f"\n{sep}\n  QUIET MODEL  (hours {sorted(QUIET_HOURS)})\n{sep}")
    tr_dl_q, va_dl_q, _, _ = _build_split_loaders(tr_ds, quiet_idx)
    print(f"[QUIET]  Train batches={len(tr_dl_q)}  Val batches={len(va_dl_q)}")

    bilstm_q  = BiLSTMForecaster(num_features=n_feat,
                                  hidden_size=64, num_layers=2,
                                  dropout=0.2,   pred_len=pred_len)
    trh_q, vah_q = train_model(bilstm_q, tr_dl_q, va_dl_q,
                                epochs=epochs, lr=LR,
                                name="BiLSTM-Attn [QUIET]")
    plot_loss_curves(trh_q, vah_q, "BiLSTM-Attn [QUIET]")

    patchtst_q = PatchTSTForecaster(context_len=ctx_len, pred_len=pred_len,
                                     d_model=64, num_heads=4, num_layers=2)
    trh_q2, vah_q2 = train_model(patchtst_q, tr_dl_q, va_dl_q,
                                  epochs=epochs, lr=LR*0.5,
                                  name="PatchTST-lite [QUIET]")
    plot_loss_curves(trh_q2, vah_q2, "PatchTST-lite [QUIET]")

    if use_oof:
        bl_oof_q, y_oof_q = generate_oof_preds(
            BiLSTMForecaster,
            dict(num_features=n_feat, hidden_size=64,
                 num_layers=2, dropout=0.2, pred_len=pred_len),
            tr_ds, n_folds=quiet_folds, lr=LR, epochs=OOF_EPOCHS,
            name="BiLSTM [QUIET]")
        pt_oof_q, _ = generate_oof_preds(
            PatchTSTForecaster,
            dict(context_len=ctx_len, pred_len=pred_len,
                 d_model=64, num_heads=4, num_layers=2),
            tr_ds, n_folds=quiet_folds, lr=LR*0.5, epochs=OOF_EPOCHS,
            name="PatchTST-lite [QUIET]")
    else:
        bl_oof_q, y_oof_q = generate_oof_preds(
            BiLSTMForecaster,
            dict(num_features=n_feat, hidden_size=64,
                 num_layers=2, dropout=0.2, pred_len=pred_len),
            tr_ds, n_folds=1, lr=LR, epochs=30, name="BiLSTM [QUIET]")
        pt_oof_q, _ = generate_oof_preds(
            PatchTSTForecaster,
            dict(context_len=ctx_len, pred_len=pred_len,
                 d_model=64, num_heads=4, num_layers=2),
            tr_ds, n_folds=1, lr=LR*0.5, epochs=30, name="PatchTST-lite [QUIET]")

    bl_va_q, y_va_q = collect_preds(bilstm_q,   va_dl)
    pt_va_q, _      = collect_preds(patchtst_q, va_dl)
    meta_q, use_pt_q = build_smart_ensemble(
        bilstm_q, patchtst_q,
        bl_oof_q, pt_oof_q, y_oof_q,
        bl_va_q,  pt_va_q,  y_va_q,
        label="QUIET")

    # ── Test-set evaluation (both ensembles on full test set) ─────────────────
    print(f"\n{sep}\n  TEST SET EVALUATION — Active vs Quiet Ensemble\n{sep}")
    bl_te_a, y_te = collect_preds(bilstm_a,   te_dl)
    pt_te_a, _    = collect_preds(patchtst_a, te_dl)
    X_te_a  = np.column_stack([bl_te_a, pt_te_a]) if use_pt_a else bl_te_a.reshape(-1,1)
    ens_a   = meta_a.predict(X_te_a)
    y_lpm   = scaler.inverse_transform(y_te)
    ens_a_lpm = scaler.inverse_transform(ens_a)

    bl_te_q, _  = collect_preds(bilstm_q,   te_dl)
    pt_te_q, _  = collect_preds(patchtst_q, te_dl)
    X_te_q  = np.column_stack([bl_te_q, pt_te_q]) if use_pt_q else bl_te_q.reshape(-1,1)
    ens_q   = meta_q.predict(X_te_q)
    ens_q_lpm = scaler.inverse_transform(ens_q)

    m_active = compute_metrics(y_lpm, ens_a_lpm, "Active Ensemble (all hrs)")
    m_quiet  = compute_metrics(y_lpm, ens_q_lpm, "Quiet Ensemble  (all hrs)")

    return dict(
        bilstm_active=bilstm_a,   patchtst_active=patchtst_a,
        meta_active=meta_a,       use_patchtst_active=use_pt_a,
        bilstm_quiet=bilstm_q,    patchtst_quiet=patchtst_q,
        meta_quiet=meta_q,        use_patchtst_quiet=use_pt_q,
        active_idx=active_idx,    quiet_idx=quiet_idx,
        metrics_active=m_active,  metrics_quiet=m_quiet,
    )


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL 4-E — BOUNDARY SMOOTHING
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def _blend_weight(hour: int, blend_hrs: int = BOUNDARY_BLEND_HRS) -> float:
    ACTIVE_START = 6
    ACTIVE_END   = 21
    if ACTIVE_START <= hour <= ACTIVE_END:
        return 1.0
    if hour < (ACTIVE_START - blend_hrs) or hour > (ACTIVE_END + blend_hrs):
        return 0.0
    if hour < ACTIVE_START:
        steps_into_ramp = hour - (ACTIVE_START - blend_hrs)
        return max(0.0, min(1.0, steps_into_ramp / blend_hrs))
    steps_past_end = hour - ACTIVE_END
    return max(0.0, min(1.0, 1.0 - steps_past_end / blend_hrs))


def _ensemble_predict_scaled(bilstm, patchtst, meta_learner,
                               use_patchtst, xb: torch.Tensor) -> np.ndarray:
    with torch.no_grad():
        bl = bilstm(xb).cpu().numpy().flatten()
        pt = patchtst(xb).cpu().numpy().flatten()
    X = np.column_stack([bl, pt]) if use_patchtst else bl.reshape(-1, 1)
    return meta_learner.predict(X)


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL 5 — MODEL PERSISTENCE
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def save_system_dual(dual: dict, scaler, tod_caps, metrics,
                      context_len, pred_len,
                      save_dir: Path = SAVE_DIR_DUAL):
    save_dir.mkdir(parents=True, exist_ok=True)
    torch.save(dual["bilstm_active"].state_dict(),   save_dir/"bilstm_active.pt")
    torch.save(dual["patchtst_active"].state_dict(), save_dir/"patchtst_active.pt")
    torch.save(dual["bilstm_quiet"].state_dict(),    save_dir/"bilstm_quiet.pt")
    torch.save(dual["patchtst_quiet"].state_dict(),  save_dir/"patchtst_quiet.pt")
    joblib.dump(dual["meta_active"],  save_dir/"meta_learner_active.pkl")
    joblib.dump(dual["meta_quiet"],   save_dir/"meta_learner_quiet.pkl")
    joblib.dump(scaler,               save_dir/"scaler.pkl")
    joblib.dump(tod_caps,             save_dir/"tod_caps.pkl")

    safe_m = {}
    for k, m in metrics.items():
        if isinstance(m, dict):
            safe_m[k] = {mk: (float(v) if isinstance(v, float) and not np.isnan(v)
                              else None)
                         for mk, v in m.items()}
        else:
            safe_m[k] = m

    config = dict(
        context_len=context_len, pred_len=pred_len,
        n_features=N_FEATURES, feature_cols=FEATURE_COLS,
        tod_clamp_quantile=TOD_CLAMP_QUANTILE,
        active_hours=sorted(ACTIVE_HOURS),
        quiet_hours=sorted(QUIET_HOURS),
        boundary_blend_hrs=BOUNDARY_BLEND_HRS,
        use_patchtst_active=dual["use_patchtst_active"],
        use_patchtst_quiet=dual["use_patchtst_quiet"],
        solution="4_dual_model_cas",
        saved_at=pd.Timestamp.now().isoformat(),
        metrics=safe_m,
    )
    with open(save_dir/"config.json","w") as f:
        json.dump(config, f, indent=2)
    print(f"\n[SAVE]  ✓ Dual-model system saved to {save_dir.resolve()}")


def load_system_dual(save_dir: Path = SAVE_DIR_DUAL) -> dict:
    with open(save_dir/"config.json") as f:
        config = json.load(f)
    ctx_len  = config["context_len"]
    pred_len = config["pred_len"]
    use_pt_a = config["use_patchtst_active"]
    use_pt_q = config["use_patchtst_quiet"]

    bilstm_a  = BiLSTMForecaster(num_features=N_FEATURES, pred_len=pred_len)
    bilstm_a.load_state_dict(torch.load(save_dir/"bilstm_active.pt",
                                         map_location=DEVICE)); bilstm_a.eval().to(DEVICE)

    patchtst_a = PatchTSTForecaster(context_len=ctx_len, pred_len=pred_len)
    patchtst_a.load_state_dict(torch.load(save_dir/"patchtst_active.pt",
                                           map_location=DEVICE)); patchtst_a.eval().to(DEVICE)

    bilstm_q  = BiLSTMForecaster(num_features=N_FEATURES, hidden_size=64,
                                  num_layers=2, dropout=0.2, pred_len=pred_len)
    bilstm_q.load_state_dict(torch.load(save_dir/"bilstm_quiet.pt",
                                         map_location=DEVICE)); bilstm_q.eval().to(DEVICE)

    patchtst_q = PatchTSTForecaster(context_len=ctx_len, pred_len=pred_len,
                                     d_model=64, num_heads=4, num_layers=2)
    patchtst_q.load_state_dict(torch.load(save_dir/"patchtst_quiet.pt",
                                           map_location=DEVICE)); patchtst_q.eval().to(DEVICE)

    meta_a   = joblib.load(save_dir/"meta_learner_active.pkl")
    meta_q   = joblib.load(save_dir/"meta_learner_quiet.pkl")
    scaler   = joblib.load(save_dir/"scaler.pkl")
    tod_caps = joblib.load(save_dir/"tod_caps.pkl")

    print(f"\n[LOAD]  ✓ Dual-model system loaded from {save_dir.resolve()}")
    print(f"[LOAD]  Context={ctx_len}h  Pred={pred_len}h")
    return dict(
        bilstm_active=bilstm_a,    patchtst_active=patchtst_a,
        meta_active=meta_a,        use_patchtst_active=use_pt_a,
        bilstm_quiet=bilstm_q,     patchtst_quiet=patchtst_q,
        meta_quiet=meta_q,         use_patchtst_quiet=use_pt_q,
        scaler=scaler, tod_caps=tod_caps,
        context_len=ctx_len, pred_len=pred_len,
        config=config,
    )


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL 5-B — MAIN DUAL-MODEL TRAINING PIPELINE
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def run_pipeline_dual(filepath,
                       context_len=CONTEXT_LEN, pred_len=PRED_LEN,
                       epochs=EPOCHS, use_oof=True,
                       deploy_mode=False) -> dict:
    pipe     = build_pipeline(filepath, context_len, pred_len,
                               deploy_mode=deploy_mode)
    scaler   = pipe["scaler"]
    tod_caps = pipe["tod_caps"]
    dual     = train_dual_models(pipe, epochs=epochs, use_oof=use_oof)
    metrics  = {
        "active_ensemble": dual["metrics_active"],
        "quiet_ensemble":  dual["metrics_quiet"],
    }
    print("\n[AUTO-SAVE]  Saving dual-model system...")
    save_system_dual(dual, scaler, tod_caps, metrics, context_len, pred_len)
    print("\n[DONE]  Dual-model pipeline complete.")
    return dict(**dual, scaler=scaler, tod_caps=tod_caps,
                pipe=pipe, metrics=metrics)


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL S4-A — TOD STATS TABLE (multi-quantile)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def build_tod_stats_table(df_clean_hourly: pd.DataFrame,
                           quantiles: list = None) -> dict:
    if quantiles is None:
        quantiles = [0.10, 0.25, 0.50, 0.75, 0.90, 0.95]
    flow = df_clean_hourly["flowrate"].dropna()
    stats = {}
    for h in range(24):
        vals = flow[flow.index.hour == h].values
        if len(vals) > 0:
            entry = {"mean": float(np.mean(vals)), "count": len(vals)}
            for q in quantiles:
                entry[f"q{int(q*100):02d}"] = float(np.quantile(vals, q))
            stats[h] = entry
        else:
            entry = {"mean": 0.0, "count": 0}
            for q in quantiles:
                entry[f"q{int(q*100):02d}"] = 0.0
            stats[h] = entry
    print(f"\n[TOD-STATS]  Multi-quantile table built from "
          f"{len(flow):,} clean hourly readings")
    return stats


def build_adaptive_tod_caps(tod_stats: dict,
                              transition_hours: set = TRANSITION_HOURS,
                              transition_q: float = TRANSITION_QUANTILE,
                              normal_q: float = NORMAL_QUANTILE) -> dict:
    caps = {}
    q_norm_key  = f"q{int(normal_q*100):02d}"
    q_trans_key = f"q{int(transition_q*100):02d}"
    for h, s in tod_stats.items():
        if h in transition_hours:
            caps[h] = s.get(q_trans_key, s["mean"])
        else:
            caps[h] = s.get(q_norm_key, s["mean"])
    print(f"\n[ADAPTIVE-CAPS]  Q{int(normal_q*100)} caps for standard hours  |  "
          f"Q{int(transition_q*100)} caps for transition hours {sorted(transition_hours)}")
    for h in sorted(transition_hours):
        print(f"  Hour {h:02d}:  Q90={tod_stats[h].get(q_norm_key,0):.4f}  "
              f"→ Q50={tod_stats[h].get(q_trans_key,0):.4f} (adaptive)")
    return caps


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL S4-B — EXTENDED ACTIVE-HOUR CONTEXT SCALING (CAS)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def compute_gap_activity_scale(df_hourly_actual: pd.DataFrame,
                                 gap_start: pd.Timestamp,
                                 gap_end:   pd.Timestamp,
                                 tod_stats:  dict,
                                 active_hours: set,
                                 pre_scan_hrs:  int = CAS_PRE_SCAN_HRS,
                                 post_scan_hrs: int = CAS_POST_SCAN_HRS,
                                 min_obs:     int   = CAS_MIN_ACTIVE_OBS,
                                 threshold:   float = ACT_SCALE_THRESHOLD,
                                 scale_floor: float = CAS_SCALE_FLOOR,
                                 scale_ceil:  float = CAS_SCALE_CEIL,
                                 verbose:     bool  = CAS_VERBOSE) -> dict:
    flow_col = "flowrate"

    def _scan_window(start_dt, end_dt, label):
        actuals, hists, obs = [], [], []
        t = start_dt
        while t <= end_dt:
            h = t.hour
            if h in active_hours and t in df_hourly_actual.index:
                actual = df_hourly_actual.loc[t, flow_col]
                if not np.isnan(actual):
                    hist_mean = tod_stats.get(h, {}).get("mean", 0.0)
                    if hist_mean > 0.001:
                        actuals.append(float(actual))
                        hists.append(float(hist_mean))
                        obs.append((t, float(actual), float(hist_mean)))
            t += pd.Timedelta(hours=1)
        return actuals, hists, obs

    pre_end   = gap_start - pd.Timedelta(hours=1)
    pre_start = gap_start - pd.Timedelta(hours=pre_scan_hrs)
    pre_actuals, pre_hists, pre_obs = _scan_window(pre_start, pre_end, "PRE")

    post_start = gap_end + pd.Timedelta(hours=1)
    post_end   = gap_end + pd.Timedelta(hours=post_scan_hrs)
    post_actuals, post_hists, post_obs = _scan_window(post_start, post_end, "POST")

    def _ratio(actuals, hists, label):
        if len(actuals) < min_obs:
            return float("nan"), f"  [{label}] Only {len(actuals)} active-hour obs (need {min_obs}) — scale=N/A"
        r = np.mean(actuals) / (np.mean(hists) + 1e-9)
        trigger = r < threshold
        msg = (f"  [{label}] {len(actuals)} active-hr obs  "
               f"actual_mean={np.mean(actuals):.4f}  "
               f"hist_mean={np.mean(hists):.4f}  "
               f"ratio={r:.4f}  "
               f"{'⚠ ANOMALY DETECTED' if trigger else '✓ Normal'}")
        return r, msg

    scale_pre,  msg_pre  = _ratio(pre_actuals,  pre_hists,  "PRE-GAP ")
    scale_post, msg_post = _ratio(post_actuals, post_hists, "POST-GAP")

    if np.isnan(scale_pre) and np.isnan(scale_post):
        scale_combined = 1.0
        triggered      = False
        msg_comb       = "  No active-hour observations found → scale=1.0 (no adjustment)"
    elif np.isnan(scale_pre):
        scale_combined = float(np.clip(scale_post, scale_floor, scale_ceil))
        triggered      = scale_post < threshold
        msg_comb       = f"  Pre unavailable → using post scale={scale_combined:.4f}"
    elif np.isnan(scale_post):
        scale_combined = float(np.clip(scale_pre, scale_floor, scale_ceil))
        triggered      = scale_pre < threshold
        msg_comb       = f"  Post unavailable → using pre scale={scale_combined:.4f}"
    else:
        raw = 0.5 * scale_pre + 0.5 * scale_post
        scale_combined = float(np.clip(raw, scale_floor, scale_ceil))
        triggered      = raw < threshold
        msg_comb       = (f"  Combined (50/50 blend): "
                          f"pre={scale_pre:.4f}  post={scale_post:.4f}  "
                          f"→ combined={scale_combined:.4f}")

    if verbose:
        print(f"\n[CAS]  Extended Activity Scale for gap {gap_start} → {gap_end}")
        print(msg_pre); print(msg_post); print(msg_comb)
        if triggered:
            print(f"  ✅ Scale applied: {scale_combined:.4f}  "
                  f"(predictions scaled to {scale_combined*100:.1f}%)")
        else:
            print(f"  → No scaling applied (household activity appears normal)")

    return dict(
        scale_pre=scale_pre, scale_post=scale_post,
        scale_combined=scale_combined, triggered=triggered,
        pre_obs=pre_obs, post_obs=post_obs,
        diagnostic_msg=f"{msg_pre} | {msg_post} | {msg_comb}",
    )


def apply_tapered_activity_scaling(rec_lpm: np.ndarray,
                                    gap_start: pd.Timestamp,
                                    scale_pre:  float,
                                    scale_post: float,
                                    taper_hrs:  int   = CAS_TAPER_HRS,
                                    scale_floor: float = CAS_SCALE_FLOOR,
                                    scale_ceil:  float = CAS_SCALE_CEIL,
                                    quiet_hours: set  = None,
                                    verbose:     bool = CAS_VERBOSE) -> np.ndarray:
    n = len(rec_lpm)
    result = rec_lpm.copy()
    n_clamped = 0
    sp = float(np.clip(scale_pre,  scale_floor, scale_ceil)) \
         if not np.isnan(scale_pre)  else 1.0
    sq = float(np.clip(scale_post, scale_floor, scale_ceil)) \
         if not np.isnan(scale_post) else 1.0
    for i in range(n):
        dt = gap_start + pd.Timedelta(hours=i)
        clock_h = dt.hour
        if quiet_hours and clock_h in quiet_hours:
            continue
        alpha   = i / max(n - 1, 1)
        scale_i = (1.0 - alpha) * sp + alpha * sq
        old = result[i]
        result[i] = max(0.0, result[i] * scale_i)
        if result[i] < old - 1e-6:
            n_clamped += 1
    if verbose and n_clamped > 0:
        reduced = (rec_lpm - result)
        print(f"[CAS-TAPER]  Scaled {n_clamped}/{n} hours  "
              f"mean reduction={reduced.mean():.4f} LPM  "
              f"max reduction={reduced.max():.4f} LPM")
    elif verbose:
        print(f"[CAS-TAPER]  No hours scaled (scale_i ≥ 1.0 for all)")
    return result


def apply_adaptive_tod_clamp(rec_lpm: np.ndarray,
                              gap_start_dt: pd.Timestamp,
                              adaptive_caps: dict) -> np.ndarray:
    clamped = rec_lpm.copy()
    n_clamped = 0
    for i, val in enumerate(rec_lpm):
        dt  = gap_start_dt + pd.Timedelta(hours=i)
        cap = adaptive_caps.get(dt.hour, float("inf"))
        if val > cap:
            clamped[i] = cap
            n_clamped += 1
    if n_clamped > 0:
        print(f"[ADAPTIVE-CLAMP]  Clamped {n_clamped}/{len(rec_lpm)} hours  "
              f"(avg reduction: {(rec_lpm - clamped).mean():.4f} LPM)")
    else:
        print(f"[ADAPTIVE-CLAMP]  No clamping needed")
    return clamped


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL S4-C — SOLUTION 4 GAP FILLER (Production)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def _build_context(df_scaled: pd.DataFrame, ref_time: pd.Timestamp,
                   context_len: int) -> torch.Tensor:
    ctx_start = ref_time - pd.Timedelta(hours=context_len-1)
    ctx_df    = df_scaled.loc[ctx_start:ref_time, FEATURE_COLS].copy()
    if len(ctx_df) < context_len:
        pad = pd.DataFrame(
            np.zeros((context_len-len(ctx_df), N_FEATURES)),
            columns=FEATURE_COLS)
        ctx_df = pd.concat([pad, ctx_df], ignore_index=True)
    x_np = np.nan_to_num(
        ctx_df.values[-context_len:].astype(np.float32), nan=0.0)
    return torch.tensor(x_np).unsqueeze(0).to(DEVICE)


@torch.no_grad()
def fill_power_failure_solution4(gap_start_dt: str,
                                   gap_end_dt:   str,
                                   raw_csv_path: str,
                                   system:       dict,
                                   tod_stats:    dict,
                                   adaptive_caps: dict,
                                   apply_cas:    bool = True,
                                   apply_clamp:  bool = True,
                                   plot:         bool = True) -> pd.DataFrame:
    bilstm_a   = system["bilstm_active"];    patchtst_a = system["patchtst_active"]
    meta_a     = system["meta_active"];      use_pt_a   = system["use_patchtst_active"]
    bilstm_q   = system["bilstm_quiet"];     patchtst_q = system["patchtst_quiet"]
    meta_q     = system["meta_quiet"];       use_pt_q   = system["use_patchtst_quiet"]
    scaler     = system["scaler"]
    ctx_len    = system["context_len"]
    pred_len   = system["pred_len"]
    active_h   = set(system["config"].get("active_hours", list(range(6,22))))
    quiet_h    = set(system["config"].get("quiet_hours",
                                           list(range(22,24)) + list(range(0,6))))

    for m in [bilstm_a, patchtst_a, bilstm_q, patchtst_q]:
        m.eval()

    device = next(bilstm_a.parameters()).device
    gs = pd.Timestamp(gap_start_dt); ge = pd.Timestamp(gap_end_dt)

    df_raw = load_raw_data(raw_csv_path)
    df_i   = interpolate_short_gaps(df_raw)
    df_h   = downsample_hourly(df_i)
    df_f   = add_cyclical_features(df_h)
    df_w   = df_f.copy()
    df_w["flowrate"]  = scaler.transform(df_w["flowrate"])
    df_w["lag_24h"]   = df_w["flowrate"].shift(24).ffill().fillna(0.0)
    df_w["lag_168h"]  = df_w["flowrate"].shift(168).ffill().fillna(0.0)
    df_h_lpm = df_h.copy()

    n_gap = len(df_w.loc[gs:ge])
    if n_gap == 0:
        raise ValueError(f"No rows between {gs} and {ge}")
    print(f"\n[S4-FILL]  Reconstructing {n_gap} hours  ({gs} → {ge})")

    # Step 1: Dual-model rolling reconstruction
    rec_scaled    = np.full(n_gap, np.nan, dtype=np.float32)
    blend_weights = np.zeros(n_gap, dtype=np.float32)
    cursor = 0
    while cursor < n_gap:
        ctx_end = gs + pd.Timedelta(hours=cursor - 1)
        ctx_s   = ctx_end - pd.Timedelta(hours=ctx_len - 1)
        ctx_df  = df_w.loc[ctx_s:ctx_end, FEATURE_COLS].copy()
        if len(ctx_df) < ctx_len:
            pad = pd.DataFrame(np.zeros((ctx_len - len(ctx_df), N_FEATURES)),
                               columns=FEATURE_COLS)
            ctx_df = pd.concat([pad, ctx_df], ignore_index=True)
        xb = torch.tensor(
            np.nan_to_num(ctx_df.values[-ctx_len:].astype(np.float32), nan=0.0)
        ).unsqueeze(0).to(device)

        ens_a = _ensemble_predict_scaled(bilstm_a, patchtst_a, meta_a, use_pt_a, xb)
        ens_q = _ensemble_predict_scaled(bilstm_q, patchtst_q, meta_q, use_pt_q, xb)

        steps = min(pred_len, n_gap - cursor)
        for j in range(steps):
            abs_hr = cursor + j
            dt     = gs + pd.Timedelta(hours=abs_hr)
            w      = _blend_weight(dt.hour)
            blended = w * ens_a[j] + (1.0 - w) * ens_q[j]
            rec_scaled[abs_hr]    = blended
            blend_weights[abs_hr] = w
            if dt in df_w.index:
                df_w.loc[dt, "flowrate"] = blended
                h_c, dw_c, mo_c = dt.hour, dt.dayofweek, dt.month
                df_w.loc[dt, "hour_sin"]  = np.sin(2*np.pi*h_c/24)
                df_w.loc[dt, "hour_cos"]  = np.cos(2*np.pi*h_c/24)
                df_w.loc[dt, "day_sin"]   = np.sin(2*np.pi*dw_c/7)
                df_w.loc[dt, "day_cos"]   = np.cos(2*np.pi*dw_c/7)
                df_w.loc[dt, "month_sin"] = np.sin(2*np.pi*mo_c/12)
                df_w.loc[dt, "month_cos"] = np.cos(2*np.pi*mo_c/12)
                l24  = dt - pd.Timedelta(hours=24)
                l168 = dt - pd.Timedelta(hours=168)
                df_w.loc[dt, "lag_24h"]  = df_w.loc[l24,  "flowrate"] if l24  in df_w.index else 0.0
                df_w.loc[dt, "lag_168h"] = df_w.loc[l168, "flowrate"] if l168 in df_w.index else 0.0
        cursor += steps

    rec_lpm_dual = scaler.inverse_transform(rec_scaled)

    # Step 2: Adaptive TOD Clamp
    if apply_clamp:
        print(f"\n[S4]  Step 2: Adaptive TOD Clamp (Q50 transition, Q90 standard)...")
        rec_lpm_clamped = apply_adaptive_tod_clamp(rec_lpm_dual, gs, adaptive_caps)
    else:
        rec_lpm_clamped = rec_lpm_dual.copy()

    # Step 3: Extended CAS
    if apply_cas:
        print(f"\n[S4]  Step 3: Extended Active-Hour Context Scaling (±{CAS_PRE_SCAN_HRS}h scan)...")
        cas_result = compute_gap_activity_scale(
            df_hourly_actual=df_h_lpm, gap_start=gs, gap_end=ge,
            tod_stats=tod_stats, active_hours=active_h,
            pre_scan_hrs=CAS_PRE_SCAN_HRS, post_scan_hrs=CAS_POST_SCAN_HRS,
            verbose=True)
        if cas_result["triggered"]:
            rec_lpm_final = apply_tapered_activity_scaling(
                rec_lpm_clamped, gs,
                scale_pre=cas_result["scale_pre"]  if not np.isnan(cas_result["scale_pre"])  else cas_result["scale_combined"],
                scale_post=cas_result["scale_post"] if not np.isnan(cas_result["scale_post"]) else cas_result["scale_combined"],
                quiet_hours=quiet_h, verbose=True)
        else:
            rec_lpm_final = rec_lpm_clamped
            cas_result["scale_combined"] = 1.0
    else:
        rec_lpm_final = rec_lpm_clamped
        cas_result = {"triggered": False, "scale_combined": 1.0,
                      "pre_obs": [], "post_obs": [],
                      "scale_pre": float("nan"), "scale_post": float("nan")}

    # ── CORRECTED: LPM·h × 12 = litres  ──────────────────────────────────────
    total_litres_dual   = float(rec_lpm_dual.sum()   * LPM_H_TO_LITRES)
    total_litres_clamp  = float(rec_lpm_clamped.sum()* LPM_H_TO_LITRES)
    total_litres_final  = float(rec_lpm_final.sum()  * LPM_H_TO_LITRES)

    print(f"\n[S4-FILL]  Dual-only mean : {rec_lpm_dual.mean():.4f} LPM  "
          f"| Total: {total_litres_dual:.4f} litres")
    print(f"[S4-FILL]  After clamp    : {rec_lpm_clamped.mean():.4f} LPM  "
          f"| Total: {total_litres_clamp:.4f} litres")
    print(f"[S4-FILL]  Final (+ CAS)  : {rec_lpm_final.mean():.4f} LPM  "
          f"| Total: {total_litres_final:.4f} litres")

    if plot:
        _plot_solution4_comparison(
            gs, ge, df_f, scaler,
            rec_lpm_dual, rec_lpm_clamped, rec_lpm_final, cas_result)

    out_idx = pd.date_range(gs, periods=n_gap, freq="1h")
    return pd.DataFrame({
        "reconstructed_lpm" : rec_lpm_final,
        "reconstructed_litres": rec_lpm_final * LPM_H_TO_LITRES,   # CORRECTED ×12
        "after_dual_lpm"    : rec_lpm_dual,
        "after_dual_litres" : rec_lpm_dual    * LPM_H_TO_LITRES,   # CORRECTED ×12
        "after_clamp_lpm"   : rec_lpm_clamped,
        "after_clamp_litres": rec_lpm_clamped * LPM_H_TO_LITRES,   # CORRECTED ×12
        "activity_scale"    : cas_result["scale_combined"],
        "cas_triggered"     : cas_result["triggered"],
        "blend_weight"      : blend_weights,
    }, index=out_idx)


def _plot_solution4_comparison(gs, ge, df_f, scaler,
                                 rec_lpm_dual, rec_lpm_clamped, rec_lpm_final,
                                 cas_result):
    vis_s = gs - pd.Timedelta(hours=48)
    vis_e = ge + pd.Timedelta(hours=24)
    vis_r = df_f.loc[vis_s:vis_e, "flowrate"].copy()
    vis_l = pd.Series(
        scaler.inverse_transform(vis_r.fillna(0).values),
        index=vis_r.index)
    n_gap = len(rec_lpm_dual)
    gap_ts = pd.date_range(gs, periods=n_gap, freq="1h")
    vis_l_masked = vis_l.copy(); vis_l_masked.loc[gs:ge] = np.nan

    def _add_boundary(s):
        for t in [gs-pd.Timedelta(hours=1), ge+pd.Timedelta(hours=1)]:
            if t in vis_l.index: s.loc[t] = vis_l.loc[t]
        return s.sort_index()

    s1 = _add_boundary(pd.Series(rec_lpm_dual,    index=gap_ts))
    s2 = _add_boundary(pd.Series(rec_lpm_clamped, index=gap_ts))
    s3 = _add_boundary(pd.Series(rec_lpm_final,   index=gap_ts))
    cas_txt = (f"CAS scale={cas_result['scale_combined']:.4f}  "
               f"({'triggered' if cas_result['triggered'] else 'not triggered'})")

    fig, axes = plt.subplots(3, 1, figsize=(17, 12), sharex=True)
    fig.suptitle(f"Solution 4 — Three-Stage Gap Filler\n{gs}  →  {ge}",
                 fontweight="bold", fontsize=14)
    combos = [
        (axes[0], s1, "#1976D2", "Dual Model (raw)",       "#BBDEFB", "Stage 1: Dual Model Only"),
        (axes[1], s2, "#FF9800", "Adaptive Clamp",         "#FFF9C4", "Stage 2: + Adaptive TOD Clamp (Q50 transition)"),
        (axes[2], s3, "#E53935", "Final (Dual+Clamp+CAS)", "#FFCDD2",
                  f"Stage 3: + Contextual Activity Scaling  [{cas_txt}]"),
    ]
    for ax, series, col, lbl, shade, title in combos:
        ax.plot(vis_l_masked.index, vis_l_masked.values,
                color="#1565C0", lw=2, label="Known Data")
        ax.plot(series.index, series.values, color=col, lw=2.5, ls="--", label=lbl)
        ax.axvspan(gs, ge, color=shade, alpha=0.4, label="Gap")
        ax.set_ylabel("LPM"); ax.legend(loc="upper right"); ax.grid(alpha=0.25)
        ax.set_title(title, fontsize=11)
    axes[-1].set_xlabel("Timestamp")
    axes[-1].xaxis.set_major_formatter(mdates.DateFormatter("%m-%d %H:%M"))
    fig.autofmt_xdate(rotation=30)
    plt.tight_layout(); plt.show()


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL S4-D — INTERNAL GAP FILLER (for evaluation)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def _run_s4_gap_filler(gs, ge, df_te, df_clean_hourly,
                        bilstm_a, patchtst_a, meta_a, use_pt_a,
                        bilstm_q, patchtst_q, meta_q, use_pt_q,
                        scaler, adaptive_caps, tod_stats, context_len,
                        active_hours, quiet_hours,
                        apply_cas=True, apply_clamp=True):
    device = next(bilstm_a.parameters()).device
    for m in [bilstm_a, patchtst_a, bilstm_q, patchtst_q]:
        m.eval()

    df_w  = df_te.copy()
    n_gap = len(df_w.loc[gs:ge])
    rec_scaled = np.full(n_gap, np.nan, dtype=np.float32)
    cursor = 0

    while cursor < n_gap:
        ctx_end = gs + pd.Timedelta(hours=cursor - 1)
        ctx_s   = ctx_end - pd.Timedelta(hours=context_len - 1)
        ctx_df  = df_w.loc[ctx_s:ctx_end, FEATURE_COLS].copy()
        if len(ctx_df) < context_len:
            pad = pd.DataFrame(
                np.zeros((context_len - len(ctx_df), N_FEATURES)),
                columns=FEATURE_COLS)
            ctx_df = pd.concat([pad, ctx_df], ignore_index=True)
        xb = torch.tensor(
            np.nan_to_num(ctx_df.values[-context_len:].astype(np.float32), nan=0.0)
        ).unsqueeze(0).to(device)

        ens_a = _ensemble_predict_scaled(bilstm_a, patchtst_a, meta_a, use_pt_a, xb)
        ens_q = _ensemble_predict_scaled(bilstm_q, patchtst_q, meta_q, use_pt_q, xb)

        steps = min(bilstm_a.pred_len, n_gap - cursor)
        for j in range(steps):
            abs_hr = cursor + j
            dt     = gs + pd.Timedelta(hours=abs_hr)
            w      = _blend_weight(dt.hour)
            blended = w * ens_a[j] + (1.0 - w) * ens_q[j]
            rec_scaled[abs_hr] = blended
            if dt in df_w.index:
                df_w.loc[dt, "flowrate"] = blended
                h_c, dw_c, mo_c = dt.hour, dt.dayofweek, dt.month
                df_w.loc[dt, "hour_sin"]  = np.sin(2*np.pi*h_c/24)
                df_w.loc[dt, "hour_cos"]  = np.cos(2*np.pi*h_c/24)
                df_w.loc[dt, "day_sin"]   = np.sin(2*np.pi*dw_c/7)
                df_w.loc[dt, "day_cos"]   = np.cos(2*np.pi*dw_c/7)
                df_w.loc[dt, "month_sin"] = np.sin(2*np.pi*mo_c/12)
                df_w.loc[dt, "month_cos"] = np.cos(2*np.pi*mo_c/12)
                l24  = dt - pd.Timedelta(hours=24)
                l168 = dt - pd.Timedelta(hours=168)
                df_w.loc[dt, "lag_24h"]  = df_w.loc[l24,  "flowrate"] if l24  in df_w.index else 0.0
                df_w.loc[dt, "lag_168h"] = df_w.loc[l168, "flowrate"] if l168 in df_w.index else 0.0
        cursor += steps

    rec_lpm = scaler.inverse_transform(rec_scaled)

    if apply_clamp:
        rec_lpm = apply_adaptive_tod_clamp(rec_lpm, gs, adaptive_caps)

    if apply_cas:
        cas = compute_gap_activity_scale(
            df_hourly_actual=df_clean_hourly,
            gap_start=gs, gap_end=ge,
            tod_stats=tod_stats, active_hours=active_hours,
            pre_scan_hrs=CAS_PRE_SCAN_HRS, post_scan_hrs=CAS_POST_SCAN_HRS,
            verbose=False)
        if cas["triggered"]:
            sp = cas["scale_pre"]  if not np.isnan(cas["scale_pre"])  else cas["scale_combined"]
            sq = cas["scale_post"] if not np.isnan(cas["scale_post"]) else cas["scale_combined"]
            rec_lpm = apply_tapered_activity_scaling(
                rec_lpm, gs, sp, sq,
                quiet_hours=quiet_hours, verbose=False)
            print(f"[CAS]  Gap {gs.date()} {gs.hour:02d}h → {ge.hour:02d}h  "
                  f"scale={cas['scale_combined']:.4f}  ⚠ TRIGGERED")
        else:
            print(f"[CAS]  Gap {gs.date()} {gs.hour:02d}h → {ge.hour:02d}h  "
                  f"scale=1.0000  (normal)")
    return rec_lpm


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL S4-E — EVALUATION TABLES  (CORRECTED: ×12 litres conversion)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def evaluate_long_gaps_solution4(result: dict,
                                   tod_stats:    dict,
                                   adaptive_caps: dict,
                                   n_gaps:     int  = 5,
                                   apply_cas:  bool = True,
                                   apply_clamp: bool = True,
                                   print_each: bool = True) -> tuple:
    """
    TABLE 1: Hourly actual vs Solution 4 predicted.
    CORRECTED conversion: LPM·h × 12 = litres
    (hourly value = mean of 12 × 5-min readings)
    """
    pipe         = result["pipe"]
    scaler       = result["scaler"]
    inject_meta  = pipe["inject_meta"]
    df_clean     = pipe["df_clean_hourly"]
    df_te        = pipe["df_test"]
    context_len  = pipe["context_len"]

    bilstm_a  = result["bilstm_active"];  patchtst_a = result["patchtst_active"]
    meta_a    = result["meta_active"];    use_pt_a   = result["use_patchtst_active"]
    bilstm_q  = result["bilstm_quiet"];   patchtst_q = result["patchtst_quiet"]
    meta_q    = result["meta_quiet"];     use_pt_q   = result["use_patchtst_quiet"]
    active_h  = ACTIVE_HOURS
    quiet_h   = QUIET_HOURS

    t_min = df_te.index.min(); t_max = df_te.index.max()
    test_pairs = [
        (s.floor("h"), e.ceil("h"))
        for s, e in zip(inject_meta["long_gap_starts"],
                        inject_meta["long_gap_ends"])
        if (s.floor("h") - pd.Timedelta(hours=context_len)) >= t_min
        and e.ceil("h") <= t_max
    ][:n_gaps]

    if not test_pairs:
        print("[EVAL-S4]  No long-gap events in test partition.")
        return pd.DataFrame(), pd.DataFrame()

    all_rows, gap_summary = [], []

    print("\n" + "╔" + "═"*76 + "╗")
    print("║  TABLE 1 — SOLUTION 4 LONG-GAP EVALUATION                         ║")
    print("║  Dual Model + Adaptive Q50/Q90 Clamp + Extended CAS               ║")
    print("║  UNIT CONVERSION: LPM·h × 12 = litres                             ║")
    print("╚" + "═"*76 + "╝")

    for g_idx, (gs, ge) in enumerate(test_pairs, 1):
        gt = df_clean.loc[gs:ge, "flowrate"].dropna()
        if len(gt) < 2: continue

        rec_lpm = _run_s4_gap_filler(
            gs, ge, df_te, df_clean,
            bilstm_a, patchtst_a, meta_a, use_pt_a,
            bilstm_q, patchtst_q, meta_q, use_pt_q,
            scaler, adaptive_caps, tod_stats, context_len,
            active_h, quiet_h,
            apply_cas=apply_cas, apply_clamp=apply_clamp)

        pred_ts = pd.date_range(gs, periods=len(rec_lpm), freq="1h")
        common  = pred_ts[pred_ts.isin(gt.index)]

        rows = []
        for ts in common:
            idx_pos   = list(pred_ts).index(ts)
            actual    = float(gt.loc[ts])
            predicted = float(rec_lpm[idx_pos])
            # CORRECTED: ×12 for litres
            actual_l    = actual    * LPM_H_TO_LITRES
            predicted_l = predicted * LPM_H_TO_LITRES
            model_tag = ("Active" if _blend_weight(ts.hour) == 1.0 else
                         "Quiet"  if _blend_weight(ts.hour) == 0.0 else "Blend")
            ae   = abs(actual - predicted)
            ae_l = abs(actual_l - predicted_l)
            direction = "↑ Over" if predicted > actual else \
                        "↓ Under" if predicted < actual else "✓ Exact"
            rows.append({
                "Gap #"             : g_idx,
                "Timestamp"         : ts,
                "Hour"              : ts.hour,
                "Model"             : model_tag,
                "Actual (LPM)"      : round(actual,    5),
                "Predicted (LPM)"   : round(predicted, 5),
                "Actual (L)"        : round(actual_l,  5),   # CORRECTED ×12
                "Predicted (L)"     : round(predicted_l, 5), # CORRECTED ×12
                "Abs Error (LPM)"   : round(ae,   5),
                "Abs Error (L)"     : round(ae_l, 5),        # CORRECTED ×12
                "Direction"         : direction,
            })

        gap_df     = pd.DataFrame(rows)
        tot_actual     = gap_df["Actual (LPM)"].sum()
        tot_pred       = gap_df["Predicted (LPM)"].sum()
        # CORRECTED: ×12 for litres
        tot_actual_l   = tot_actual * LPM_H_TO_LITRES
        tot_pred_l     = tot_pred   * LPM_H_TO_LITRES
        tot_ae         = gap_df["Abs Error (LPM)"].sum()
        tot_ae_l       = tot_ae * LPM_H_TO_LITRES
        y_t  = gap_df["Actual (LPM)"].values
        y_p  = gap_df["Predicted (LPM)"].values
        rmse = float(np.sqrt(mean_squared_error(y_t, y_p)))
        mae  = float(mean_absolute_error(y_t, y_p))
        ss   = float(np.sum((y_t - y_t.mean())**2))
        r2   = float(r2_score(y_t, y_p)) if (len(y_t) >= 5 and ss >= 1e-8) \
               else float("nan")
        r2s  = f"{r2:.4f}" if not np.isnan(r2) else "N/A"

        gap_summary.append({
            "Gap #"                    : g_idx,
            "Start"                    : str(gs),
            "End"                      : str(ge),
            "Valid Hours"              : len(common),
            "Total Actual (LPM·h)"    : round(tot_actual,   4),
            "Total Predicted (LPM·h)" : round(tot_pred,     4),
            "Total Actual (L)"         : round(tot_actual_l, 4),  # CORRECTED ×12
            "Total Predicted (L)"      : round(tot_pred_l,   4),  # CORRECTED ×12
            "Difference (LPM·h)"      : round(tot_pred   - tot_actual,   4),
            "Difference (L)"           : round(tot_pred_l - tot_actual_l, 4),  # CORRECTED ×12
            "Total Abs Error (LPM·h)" : round(tot_ae,   4),
            "Total Abs Error (L)"      : round(tot_ae_l, 4),  # CORRECTED ×12
            "RMSE"                     : round(rmse, 5),
            "MAE"                      : round(mae,  5),
            "R²"                       : r2s,
        })

        if print_each:
            pd.set_option("display.float_format", "{:.5f}".format)
            pd.set_option("display.max_columns", 12)
            pd.set_option("display.width", 140)
            print(f"\n  {'─'*74}")
            print(f"  GAP #{g_idx}  |  {gs}  →  {ge}  ({len(common)} valid hours)")
            print(f"  Model routing: "
                  f"Active={(gap_df['Model']=='Active').sum()}h  "
                  f"Quiet={(gap_df['Model']=='Quiet').sum()}h  "
                  f"Blend={(gap_df['Model']=='Blend').sum()}h")
            print(f"  {'─'*74}")
            print(gap_df.to_string(index=False))
            print(f"\n  ┌─ TOTALS {'─'*60}┐")
            print(f"  │  Total Actual    : {tot_actual:>10.5f} LPM·h  = {tot_actual_l:>10.5f} litres  │")
            print(f"  │  Total Predicted : {tot_pred:>10.5f} LPM·h  = {tot_pred_l:>10.5f} litres  (S4) │")
            print(f"  │  Difference      : {tot_pred-tot_actual:>+10.5f} LPM·h  = {tot_pred_l-tot_actual_l:>+10.5f} litres  │")
            print(f"  │  RMSE={rmse:.5f}  MAE={mae:.5f}  R²={r2s}" + " "*10 + "│")
            print(f"  └{'─'*70}┘")

        all_rows.extend(rows)

    full_df    = pd.DataFrame(all_rows)
    summary_df = pd.DataFrame(gap_summary)

    print("\n\n" + "╔" + "═"*76 + "╗")
    print("║  SOLUTION 4 LONG-GAP SUMMARY  (CORRECTED: ×12 litres)            ║")
    print("╚" + "═"*76 + "╝")
    print(summary_df.to_string(index=False))

    grand_actual   = summary_df["Total Actual (LPM·h)"].sum()
    grand_pred     = summary_df["Total Predicted (LPM·h)"].sum()
    grand_diff     = grand_pred - grand_actual
    grand_pct      = abs(grand_diff) / (grand_actual + 1e-6) * 100
    # CORRECTED: ×12 for litres
    grand_actual_l = grand_actual * LPM_H_TO_LITRES
    grand_pred_l   = grand_pred   * LPM_H_TO_LITRES
    grand_diff_l   = grand_diff   * LPM_H_TO_LITRES

    print(f"\n  {'═'*62}")
    print(f"  GRAND TOTAL  ({len(summary_df)} long-gap events)  [Solution 4]")
    print(f"  {'─'*62}")
    print(f"  Total Actual Usage    : {grand_actual:.4f} LPM·h  = {grand_actual_l:.4f} litres")
    print(f"  Total Predicted Usage : {grand_pred:.4f} LPM·h  = {grand_pred_l:.4f} litres")
    print(f"  Overall Difference    : {grand_diff:+.4f} LPM·h  = {grand_diff_l:+.4f} litres  "
          f"({'over-estimate' if grand_diff>0 else 'under-estimate'})")
    print(f"  Overall % Error       : {grand_pct:.2f}%")
    print(f"  {'═'*62}")

    if len(summary_df) > 0:
        fig, ax = plt.subplots(figsize=(14, 6))
        fig.suptitle("Solution 4 — Total Usage per Gap: Actual vs Predicted (litres)",
                     fontsize=14, fontweight="bold")
        x = np.arange(len(summary_df)); w = 0.35
        ax.bar(x - w/2, summary_df["Total Actual (L)"],    w,
               label="Actual (L)",       color="#1976D2", alpha=0.85)
        ax.bar(x + w/2, summary_df["Total Predicted (L)"], w,
               label="S4 Predicted (L)", color="#E53935", alpha=0.85)
        ax.set_xticks(x)
        ax.set_xticklabels([f"Gap {i}" for i in summary_df["Gap #"]])
        ax.set_ylabel("Total Usage (litres)")
        ax.legend(); ax.grid(axis="y", alpha=0.3)
        plt.tight_layout(); plt.show()

        # Scatter: actual vs predicted (LPM)
        fig2, ax2 = plt.subplots(figsize=(8, 7))
        colors = full_df["Model"].map(
            {"Active": "#1976D2", "Quiet": "#7B1FA2", "Blend": "#FF9800"})
        ax2.scatter(full_df["Actual (LPM)"], full_df["Predicted (LPM)"],
                    c=colors, alpha=0.7, s=60, edgecolors="white", lw=0.5)
        lo = min(full_df["Actual (LPM)"].min(), full_df["Predicted (LPM)"].min()) - 0.02
        hi = max(full_df["Actual (LPM)"].max(), full_df["Predicted (LPM)"].max()) + 0.02
        ax2.plot([lo, hi], [lo, hi], "r--", lw=1.5, label="Perfect prediction")
        from matplotlib.lines import Line2D
        ax2.legend(handles=[
            Line2D([0],[0], marker='o', color='w', markerfacecolor='#1976D2',
                   markersize=8, label='Active Model'),
            Line2D([0],[0], marker='o', color='w', markerfacecolor='#7B1FA2',
                   markersize=8, label='Quiet Model'),
            Line2D([0],[0], marker='o', color='w', markerfacecolor='#FF9800',
                   markersize=8, label='Blended'),
            Line2D([0],[0], color='r', ls='--', label='Perfect'),
        ])
        ax2.set_xlabel("Actual (LPM)"); ax2.set_ylabel("Predicted (LPM)")
        ax2.set_title("Hourly Actual vs Predicted — by Model Type", fontweight="bold")
        ax2.grid(alpha=0.3); ax2.set_xlim([lo, hi]); ax2.set_ylim([lo, hi])
        plt.tight_layout(); plt.show()

    return full_df, summary_df


def evaluate_short_gaps_table(result: dict, n_gaps: int = 10,
                               print_table: bool = True) -> tuple:
    """
    TABLE 2: Short-gap cubic-spline interpolation.
    CORRECTED conversion: LPM·5min × 1 = litres
    (each 5-min reading represents 1/12 of an hour; ×12/12 = ×1)
    """
    pipe       = result["pipe"]
    df_raw     = pipe["df_raw"]
    df_interp  = pipe["df_interpolated"]
    inj_meta   = pipe["inject_meta"]
    mask       = inj_meta["mask"]

    print("\n\n" + "╔" + "═"*76 + "╗")
    print("║  TABLE 2 — SHORT-GAP (≤ 15 MIN INTERPOLATION) EVALUATION          ║")
    print("║  UNIT CONVERSION: LPM·5min × 1 = litres                           ║")
    print("╚" + "═"*76 + "╝")

    vals_raw    = df_raw["flowrate"].values
    vals_interp = df_interp["flowrate"].values
    idx         = df_raw.index; n = len(vals_raw)

    runs, in_gap, gs_i = [], False, 0
    for i in range(n):
        if mask[i] and not in_gap:  in_gap, gs_i = True, i
        elif not mask[i] and in_gap: runs.append((gs_i, i-1)); in_gap = False
    if in_gap: runs.append((gs_i, n-1))
    short_runs = [(s, e) for s, e in runs if (e-s+1) <= 3][:n_gaps]

    if not short_runs:
        print("[EVAL]  No short-gap events found.")
        return pd.DataFrame(), pd.DataFrame()

    all_rows, gap_summary = [], []
    for g_idx, (s, e) in enumerate(short_runs, 1):
        gap_len = e - s + 1; rows = []
        for pos in range(s, e+1):
            ts          = idx[pos]
            actual      = float(vals_raw[pos])
            predicted   = float(vals_interp[pos])
            # CORRECTED: LPM·5min × 1 = litres
            actual_l    = actual    * LPM_5MIN_TO_LITRES
            predicted_l = predicted * LPM_5MIN_TO_LITRES
            ae          = abs(actual - predicted)
            ae_l        = abs(actual_l - predicted_l)
            direction   = "↑ Over" if predicted > actual else \
                          "↓ Under" if predicted < actual else "✓ Exact"
            rows.append({
                "Gap #"              : g_idx,
                "Timestamp"          : ts,
                "Time"               : ts.strftime("%H:%M"),
                "Actual (LPM)"       : round(actual,    5),
                "Interpolated (LPM)" : round(predicted, 5),
                "Actual (L)"         : round(actual_l,  5),   # CORRECTED ×1
                "Interpolated (L)"   : round(predicted_l, 5), # CORRECTED ×1
                "Abs Error (LPM)"    : round(ae,   5),
                "Abs Error (L)"      : round(ae_l, 5),        # CORRECTED ×1
                "Direction"          : direction,
            })
        gap_df     = pd.DataFrame(rows)
        tot_actual     = gap_df["Actual (LPM)"].sum()
        tot_pred       = gap_df["Interpolated (LPM)"].sum()
        # CORRECTED: ×1 for 5-min litres
        tot_actual_l   = tot_actual * LPM_5MIN_TO_LITRES
        tot_pred_l     = tot_pred   * LPM_5MIN_TO_LITRES
        gap_summary.append({
            "Gap #"                    : g_idx,
            "Start"                    : str(idx[s]),
            "End"                      : str(idx[e]),
            "Points"                   : gap_len,
            "Duration"                 : f"{gap_len*5} min",
            "Total Actual (LPM·5min)"  : round(tot_actual,   5),
            "Total Interp (LPM·5min)"  : round(tot_pred,     5),
            "Total Actual (L)"          : round(tot_actual_l, 5),  # CORRECTED ×1
            "Total Interp (L)"          : round(tot_pred_l,   5),  # CORRECTED ×1
            "Difference (LPM·5min)"    : round(tot_pred   - tot_actual,   5),
            "Difference (L)"            : round(tot_pred_l - tot_actual_l, 5),  # CORRECTED ×1
        })
        if print_table:
            print(f"\n  {'─'*74}")
            print(f"  SHORT GAP #{g_idx}  |  {idx[s]}  →  {idx[e]}  "
                  f"({gap_len} pts = {gap_len*5} min)")
            print(f"  {'─'*74}")
            print(gap_df.to_string(index=False))
        all_rows.extend(rows)

    short_full_df = pd.DataFrame(all_rows)
    summary_df    = pd.DataFrame(gap_summary)

    print("\n\n" + "╔" + "═"*76 + "╗")
    print("║  SHORT-GAP SUMMARY  (CORRECTED: LPM·5min × 1 = litres)           ║")
    print("╚" + "═"*76 + "╝")
    print(summary_df.to_string(index=False))

    grand_actual   = summary_df["Total Actual (LPM·5min)"].sum()
    grand_pred     = summary_df["Total Interp (LPM·5min)"].sum()
    grand_diff     = grand_pred - grand_actual
    grand_pct      = abs(grand_diff) / (grand_actual + 1e-6) * 100
    # CORRECTED ×1 for 5-min
    grand_actual_l = grand_actual * LPM_5MIN_TO_LITRES
    grand_pred_l   = grand_pred   * LPM_5MIN_TO_LITRES
    grand_diff_l   = grand_diff   * LPM_5MIN_TO_LITRES

    print(f"\n  Overall Abs Difference: {grand_diff:+.5f} LPM·5min  "
          f"= {grand_diff_l:+.5f} litres  ({grand_pct:.2f}%)")
    return short_full_df, summary_df


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  CELL S4-F — MASTER EVALUATION ENTRY POINT  (CORRECTED conversions)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def run_gap_evaluation_solution4(result: dict,
                                   n_long_gaps:  int  = 5,
                                   n_short_gaps: int  = 10,
                                   apply_cas:    bool = True,
                                   apply_clamp:  bool = True) -> dict:
    """
    Master evaluation for Solution 4.
    CORRECTED unit conversion throughout:
      - Hourly  : LPM·h     × 12 = litres
      - 5-minute: LPM·5min  ×  1 = litres
    Steps:
        1. Build multi-quantile TOD stats + adaptive caps
        2. TABLE 1: Long-gap hourly evaluation
        3. TABLE 2: Short-gap interpolation
        4. Combined final summary
    """
    pipe     = result["pipe"]
    df_clean = pipe["df_clean_hourly"]

    sep = "="*68
    print(f"\n{sep}\n  SOLUTION 4 — Building TOD stats\n{sep}")
    tod_stats     = build_tod_stats_table(df_clean)
    adaptive_caps = build_adaptive_tod_caps(tod_stats)

    try:
        SAVE_DIR_DUAL.mkdir(parents=True, exist_ok=True)
        joblib.dump(tod_stats,     SAVE_DIR_DUAL/"tod_stats.pkl")
        joblib.dump(adaptive_caps, SAVE_DIR_DUAL/"adaptive_caps.pkl")
        print(f"[S4]  tod_stats.pkl + adaptive_caps.pkl saved to {SAVE_DIR_DUAL}")
    except Exception as e:
        print(f"[S4]  Could not save stats: {e}")

    print(f"\n{sep}\n  SOLUTION 4 — Long-gap evaluation\n{sep}")
    long_full_df, s4_summary = evaluate_long_gaps_solution4(
        result, tod_stats, adaptive_caps,
        n_gaps=n_long_gaps,
        apply_cas=apply_cas,
        apply_clamp=apply_clamp,
        print_each=True)

    short_full_df, short_summary = evaluate_short_gaps_table(
        result, n_gaps=n_short_gaps, print_table=True)

    # ── Final combined summary ─────────────────────────────────────────────────
    print("\n\n" + "╔" + "═"*76 + "╗")
    print("║  FINAL COMBINED SUMMARY — Solution 4                              ║")
    print("║  Dual Model + Adaptive TOD Clamp + Extended CAS                   ║")
    print("║  CORRECTED: LPM·h × 12 = litres  |  LPM·5min × 1 = litres        ║")
    print("╚" + "═"*76 + "╝")

    if len(s4_summary) > 0:
        la    = s4_summary["Total Actual (LPM·h)"].sum()
        lp    = s4_summary["Total Predicted (LPM·h)"].sum()
        # CORRECTED ×12
        la_l  = la * LPM_H_TO_LITRES
        lp_l  = lp * LPM_H_TO_LITRES
        lp2   = abs(lp-la)/(la+1e-6)*100
        print(f"\n  Long-gap reconstruction")
        print(f"    Events    : {len(s4_summary)}")
        print(f"    Actual    : {la:.4f} LPM·h  =  {la_l:.4f} litres")
        print(f"    Predicted : {lp:.4f} LPM·h  =  {lp_l:.4f} litres")
        print(f"    Difference: {lp-la:+.4f} LPM·h  =  {(lp-la)*LPM_H_TO_LITRES:+.4f} litres  "
              f"({'over' if lp > la else 'under'})")
        print(f"    % Error   : {lp2:.2f}%")

    if len(short_summary) > 0:
        sa    = short_summary["Total Actual (LPM·5min)"].sum()
        sp    = short_summary["Total Interp (LPM·5min)"].sum()
        # CORRECTED ×1
        sa_l  = sa * LPM_5MIN_TO_LITRES
        sp_l  = sp * LPM_5MIN_TO_LITRES
        sp2   = abs(sp-sa)/(sa+1e-6)*100
        print(f"\n  Short-gap (cubic spline)")
        print(f"    Events    : {len(short_summary)}")
        print(f"    Actual    : {sa:.5f} LPM·5min  =  {sa_l:.5f} litres")
        print(f"    Predicted : {sp:.5f} LPM·5min  =  {sp_l:.5f} litres")
        print(f"    Difference: {sp-sa:+.5f} LPM·5min  =  {(sp-sa)*LPM_5MIN_TO_LITRES:+.5f} litres")
        print(f"    % Error   : {sp2:.2f}%")

    # CORRECTED note
    print(f"\n  NOTE: Hourly LPM·h × {LPM_H_TO_LITRES} = litres  "
          f"(hourly value = mean of 12 × 5-min readings)")
    print(f"  NOTE: 5-min LPM·5min × {LPM_5MIN_TO_LITRES} = litres  "
          f"(each 5-min reading is 1/12 of one hour)")

    # CSV export
    if len(long_full_df) > 0:
        long_full_df.to_csv("long_gap_hourly_detail_s4.csv",  index=False)
        s4_summary.to_csv("long_gap_summary_s4.csv",           index=False)
        print("\n[SAVED]  long_gap_hourly_detail_s4.csv")
        print("[SAVED]  long_gap_summary_s4.csv")
    if len(short_full_df) > 0:
        short_full_df.to_csv("short_gap_5min_detail.csv",  index=False)
        short_summary.to_csv("short_gap_summary.csv",       index=False)
        print("[SAVED]  short_gap_5min_detail.csv")
        print("[SAVED]  short_gap_summary.csv")

    print("\n  In Google Colab: folder icon (left sidebar) → /content/\n")

    return dict(
        tod_stats=tod_stats, adaptive_caps=adaptive_caps,
        long_full=long_full_df,   long_summary=s4_summary,
        short_full=short_full_df, short_summary=short_summary,
    )


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# ██  ENTRY POINT
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

if __name__ == "__main__":

    DATA_PATH = "colombo_house_water_usage_2024.csv"
    # from google.colab import drive; drive.mount('/content/drive')
    # DATA_PATH = "/content/drive/MyDrive/colombo_house_water_usage_2024.csv"

    # ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
    # STEP A: Train dual models (~4–5 hours on Colab GPU)
    # ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
    result = run_pipeline_dual(
        filepath    = DATA_PATH,
        context_len = CONTEXT_LEN,
        pred_len    = PRED_LEN,
        epochs      = EPOCHS,
        use_oof     = True,
        deploy_mode = False,
    )

    # ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
    # STEP B: Solution 4 evaluation
    # ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
    eval4 = run_gap_evaluation_solution4(
        result,
        n_long_gaps  = 5,
        n_short_gaps = 10,
        apply_cas    = True,
        apply_clamp  = True,
    )

    # ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
    # STEP C: Production inference
    # ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
    system        = load_system_dual()
    tod_stats     = eval4["tod_stats"]
    adaptive_caps = eval4["adaptive_caps"]

    # Example: overnight gap
    imputed = fill_power_failure_solution4(
        gap_start_dt  = "2024-11-19 00:00",
        gap_end_dt    = "2024-11-19 08:00",
        raw_csv_path  = DATA_PATH,
        system        = system,
        tod_stats     = tod_stats,
        adaptive_caps = adaptive_caps,
        apply_cas     = True,
        apply_clamp   = True,
        plot          = True,
    )
    print("\nFinal imputed gap (Solution 4) — with corrected litre columns:")
    print(imputed.to_string())

    # Example: daytime gap
    imputed_day = fill_power_failure_solution4(
        gap_start_dt  = "2024-11-21 17:00",
        gap_end_dt    = "2024-11-22 01:00",
        raw_csv_path  = DATA_PATH,
        system        = system,
        tod_stats     = tod_stats,
        adaptive_caps = adaptive_caps,
        apply_cas     = True,
        apply_clamp   = True,
        plot          = True,
    )
    print("\nFinal imputed gap daytime (Solution 4) — with corrected litre columns:")
    print(imputed_day.to_string())

    # ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
    # STEP D: Backup to Google Drive (optional)
    # ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
    """
    from google.colab import drive
    drive.mount('/content/drive')
    import shutil
    shutil.copytree('/content/smart_meter_models_dual',
                    '/content/drive/MyDrive/smart_meter_models_dual',
                    dirs_exist_ok=True)
    print("✅ All models + configs backed up to Google Drive")
    """
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def simulate_custom_gap(start_dt, end_dt, raw_csv_path, result_dict, system_dict, tod_stats, adaptive_caps):
    """
    Simulates a missing data gap, predicts the usage using Solution 4,
    and compares it with the actual ground truth in Litres.
    """
    # 1. සැබෑ දත්ත (Ground Truth) ලබා ගැනීම
    pipe = result_dict["pipe"]
    df_clean = pipe["df_clean_hourly"]

    gs = pd.Timestamp(start_dt)
    ge = pd.Timestamp(end_dt)

    actual_data = df_clean.loc[gs:ge, "flowrate"].dropna()

    if len(actual_data) == 0:
        print(f"❌ Error: {gs} සහ {ge} අතර කාලයේ සසඳා බැලීමට සැබෑ දත්ත (Ground truth) නොමැත.")
        return None

    print(f"\n⏳ {gs} සිට {ge} දක්වා කාලය Missing Data ලෙස සලකා ගණනය කරමින් පවතී...")

    # 2. ආකෘතිය හරහා පුරෝකථනය කිරීම (Solution 4 Inference)
    # මෙහිදී plot=False කර ඇත (අවශ්‍ය නම් True කළ හැක)
    imputed_df = fill_power_failure_solution4(
        gap_start_dt=start_dt,
        gap_end_dt=end_dt,
        raw_csv_path=raw_csv_path,
        system=system_dict,
        tod_stats=tod_stats,
        adaptive_caps=adaptive_caps,
        apply_cas=True,
        apply_clamp=True,
        plot=False
    )

    # 3. සැබෑ දත්ත සහ පුරෝකථනය කළ දත්ත එකට ගැලපීම (Aligning Data)
    common_index = imputed_df.index.intersection(actual_data.index)

    actual_lpm = actual_data.loc[common_index].values
    pred_lpm = imputed_df.loc[common_index, "reconstructed_lpm"].values

    # ඒකක පරිවර්තනය: පැයට LPM අගය x 12 = මුළු ලීටර් ප්‍රමාණය (Litres)
    LPM_H_TO_LITRES = 12
    actual_litres = actual_lpm * LPM_H_TO_LITRES
    pred_litres = pred_lpm * LPM_H_TO_LITRES

    # 4. දෝෂ ප්‍රතිශත ගණනය කිරීම (Calculating Metrics in Litres)
    rmse = np.sqrt(mean_squared_error(actual_litres, pred_litres))
    mae = mean_absolute_error(actual_litres, pred_litres)

    # R2 Score එක ගණනය කළ හැක්කේ දත්ත ලක්ෂ්‍ය 2කට වඩා ඇත්නම් පමණි
    r2 = r2_score(actual_litres, pred_litres) if len(actual_litres) >= 2 else float('nan')
    gap_hours = len(common_index)

    # 5. අවසාන වගුව නිර්මාණය කිරීම (Tabular Format)
    comparison_df = pd.DataFrame({
        "Timestamp": common_index.strftime('%Y-%m-%d %H:%M'),
        "Actual (Liters)": actual_litres,
        "Predicted (Liters)": pred_litres,
        "Abs Error (Liters)": np.abs(actual_litres - pred_litres)
    })

    # 6. ප්‍රතිඵල තිරයේ පෙන්වීම
    print("\n" + "═"*65)
    print(" 📊 CUSTOM GAP SIMULATION RESULTS (ඒකකය: ලීටර් / Liters)")
    print("═"*65)
    print(f" 🔹 Gap Period       : {gs}  to  {ge}")
    print(f" 🔹 Gap Duration     : {gap_hours} Hours")
    print(f" 🔹 Total Actual     : {np.sum(actual_litres):.2f} Liters")
    print(f" 🔹 Total Predicted  : {np.sum(pred_litres):.2f} Liters")
    print("-" * 65)
    print(f" 📉 RMSE (Liters)    : {rmse:.4f}")
    print(f" 📉 MAE  (Liters)    : {mae:.4f}  (Mean Absolute Error)")
    print(f" 📈 R² Score         : {r2:.4f}")
    print("═" * 65)

    print("\n 📑 වගුගත කළ දත්ත (TABULAR DATA):")
    pd.set_option('display.float_format', '{:.4f}'.format)
    print(comparison_df.to_string(index=False))
    print("\n" + "═"*65)

    return comparison_df

# =====================================================================
# 🚀 භාවිත කරන ආකාරය (How to Use)
# ඔබට අවශ්‍ය Time Range එක මෙතැනින් වෙනස් කරන්න (YYYY-MM-DD HH:MM)
# =====================================================================

START_TIME = "2024-05-10 08:00"  # හිඩැස පටන් ගන්නා වෙලාව
END_TIME   = "2024-05-10 18:00"  # හිඩැස අවසන් වන වෙලාව

# පහත විචල්‍යයන් (Variables) ඔබගේ ප්‍රධාන කේතය run කළ පසු මතකයේ ඇත
custom_simulation_table = simulate_custom_gap(
    start_dt=START_TIME,
    end_dt=END_TIME,
    raw_csv_path=DATA_PATH,       # මුල් කේතයේ ඇති CSV නම
    result_dict=result,           # මුල් කේතයෙන් ආ result dictionary එක
    system_dict=system,           # load කරගත් AI මාදිලි
    tod_stats=tod_stats,          # Time of Day stats
    adaptive_caps=adaptive_caps   # Clamping limits
)
```
